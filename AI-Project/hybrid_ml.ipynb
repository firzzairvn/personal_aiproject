```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('../dataset/google_stock.csv')
df
```

```python
df.info()
```

**Dataset Info** <br>
Dataset berasal dari data saham Alphabet Inc. / Google LLC dengan kode saham GOOG

Date = Setiap tanggal dari masing-masing data <br>
Open = Harga buka ditiap hari nya <br>
Close = Harga tutup (harga terakhir dihari itu) <br>
High = Harga tertiggi <br>
Low = Harga Terendah <br>
Volume = Banyaknya volume transaksi


```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df[['Close']] #Disini fokus pada kolom data close yang ada di dataset sebagai acuan data
```

```python
from sklearn.preprocessing import MinMaxScaler #MinMaxScaler digunakan untuk normalisasi data
scaler = MinMaxScaler(feature_range=(0, 1)) #Normalisasi data dalam range 0 - 1 
df['Close'] = scaler.fit_transform(df[['Close']])
```

```python
import numpy as np

def sequence_hari(df, seq_length=180):
    X = np.array([df[i:i+seq_length] for i in range(len(df) - seq_length)])
    y = np.array(df[seq_length:])
    return X, y

seq_length = 180 #Mengambil data untuk 180 Hari
X, y = sequence_hari(df['Close'].values, seq_length)
```

```python
train_size = int(len(X) * 0.8 ) #Training data di 80% dari total data

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(X_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=100))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=15, batch_size =32)
```

```python
df['Lagging_1'] = df['Close'].shift(1)
df['Lagging_2'] = df['Close'].shift(2)
df['Lagging_3'] = df['Close'].shift(3)
df['Lagging_4'] = df['Close'].shift(4)
df['Lagging_5'] = df['Close'].shift(5)
df = df.dropna()
```

```python
X_linReg = df[['Lagging_1','Lagging_2','Lagging_3', 'Lagging_4', 'Lagging_5']]
y_linReg = df['Close']
X_trainLinReg, X_testLinReg = X_linReg[:train_size], X_linReg[train_size:]
y_trainLinReg, y_testLinReg = y_linReg[:train_size], y_linReg[train_size:]
```

```python
from sklearn.linear_model import LinearRegression
model_linReg = LinearRegression()
model_linReg.fit(X_trainLinReg, y_trainLinReg)
```

```python
train_score = model_linReg.score(X_trainLinReg, y_trainLinReg)
test_score = model_linReg.score(X_testLinReg, y_testLinReg)
print(f"Training Score: {train_score}")
print(f"Test Score: {test_score}")
```

```python
X_testLSTM = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
LSTM_Prediction = model.predict(X_testLSTM)
LSTM_Prediction = scaler.inverse_transform(LSTM_Prediction) #Mengembalikan nilai harga ke rentang sebenarnya setelah dilakukan MinMaxScaler 0 - 1
```

```python
linReg_Prediction = model_linReg.predict(X_testLinReg)
linReg_Prediction = scaler.inverse_transform(linReg_Prediction.reshape(-1, 1))
```

```python
LSTM_Prediction = LSTM_Prediction[:min(len(LSTM_Prediction), len(linReg_Prediction))]
linReg_Prediction = linReg_Prediction[:min(len(LSTM_Prediction), len(linReg_Prediction))]

final_prediction = (0.7 * LSTM_Prediction) + (0.3 * linReg_Prediction)
```

```python
LSTM_Future_Prediction = []
last_seq = X[-1].reshape(1, seq_length, 1)

for _ in range(15):
    LSTM_Pred = model.predict(last_seq)[0, 0] 
    LSTM_Future_Prediction.append(LSTM_Pred)
    
    last_seq = np.roll(last_seq, -1, axis=1)  # Geser sequence ke kiri
    last_seq[0, -1, 0] = LSTM_Pred  # Update dengan prediksi baru

# Konversi list hasil prediksi ke dalam array dan inverse transform
LSTM_Future_Prediction = np.array(LSTM_Future_Prediction).reshape(-1, 1)
LSTM_Future_Prediction = scaler.inverse_transform(LSTM_Future_Prediction)
```

```python
from collections import deque
recent_data = deque(df['Close'].values[-5:], maxlen=5)

linReg_Future_Prediction = []

for _ in range(15):
    linReg_Pred = model_linReg.predict(np.array(recent_data).reshape(1, -1))[0]
    linReg_Future_Prediction.append(linReg_Pred)
    recent_data.append(linReg_Pred)

# Konversi list hasil prediksi ke dalam array dan inverse transform
linReg_Future_Prediction = np.array(linReg_Future_Prediction).reshape(-1, 1)
linReg_Future_Prediction = scaler.inverse_transform(linReg_Future_Prediction)
```

```python
final_Future_Prediction = (0.7 * LSTM_Future_Prediction) + (0.3 * linReg_Future_Prediction)
```

```python
future_dates = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=15)
predictions_df = pd.DataFrame({
    'Date': future_dates,
    'LSTM Predictions': LSTM_Future_Prediction.flatten(),
    'Linear Regression Predictions': linReg_Future_Prediction.flatten(),
    'Hybrid Model Predictions': final_Future_Prediction.flatten()
})
print(predictions_df)
```
